## Data Preprocessing

### Subtask:
Preprocess both the training and testing datasets. This involves handling missing values (e.g., imputation), encoding categorical features (e.g., one-hot encoding), and feature scaling if necessary, to prepare the data for model training.


**Reasoning**:
First, I'll check for missing values in the test dataset to ensure data quality. Then, I'll identify the categorical columns and apply one-hot encoding to both the training and testing datasets, taking care to align their columns to maintain consistency for model training. I will separate the target variable and prepare the final feature sets for both datasets, displaying their heads for verification.



```python
print("1. Missing values per column in test_df:")
print(test_df.isnull().sum())

# Identify categorical columns
categorical_cols = ['gender', 'marital_status', 'education_level', 'employment_status', 'loan_purpose', 'grade_subgrade']

# Store test_ids for submission file later
test_ids = test_df['id']

# Drop 'id' and 'loan_paid_back' from train_df for preprocessing features
X = train_df.drop(['id', 'loan_paid_back'], axis=1)
y_train = train_df['loan_paid_back']

# Drop 'id' from test_df for preprocessing features
X_test = test_df.drop('id', axis=1)

# Apply one-hot encoding to categorical columns for both X and X_test
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True)

# Align columns - crucial to ensure both datasets have the same columns in the same order
# Use 'outer' join and fill_value=0 to handle potential discrepancies in categories (e.g., a category exists in train but not test, or vice-versa)
X_train_final, X_test_final = X_encoded.align(X_test_encoded, join='outer', axis=1, fill_value=0)

print("\n7. Head of preprocessed X_train:")
print(X_train_final.head())

print("\nHead of y_train:")
print(y_train.head())

print("\nHead of preprocessed X_test:")
print(X_test_final.head())
```

**Reasoning**:
Given that missing values have been handled and categorical features encoded, the next step in preprocessing, as outlined in the subtask, is to apply feature scaling to the numerical columns to ensure consistent ranges for model training.



```python
from sklearn.preprocessing import StandardScaler

# Identify numerical columns for scaling
numerical_cols = ['annual_income', 'debt_to_income_ratio', 'credit_score', 'loan_amount', 'interest_rate']

# Initialize StandardScaler
scaler = StandardScaler()

# Apply scaling to numerical columns in X_train_final
X_train_final[numerical_cols] = scaler.fit_transform(X_train_final[numerical_cols])

# Apply scaling to numerical columns in X_test_final using the SAME scaler fitted on X_train
X_test_final[numerical_cols] = scaler.transform(X_test_final[numerical_cols])

print("Head of X_train_final after scaling numerical features:")
print(X_train_final.head())

print("\nHead of X_test_final after scaling numerical features:")
print(X_test_final.head())
```

## Train Prediction Model

### Subtask:
Train a machine learning classification model on the preprocessed training data to predict loan payback. A suitable model will be chosen based on the data characteristics identified during EDA.


**Reasoning**:
To train the prediction model as per the subtask, I will first import `LogisticRegression`, instantiate it with `random_state=42`, `class_weight='balanced'`, and `max_iter=1000` to handle class imbalance and ensure convergence, then fit the model using the preprocessed training data.



```python
from sklearn.linear_model import LogisticRegression

# Instantiate LogisticRegression model with specified parameters
model = LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000)

# Train the model
model.fit(X_train_final, y_train)

print("Logistic Regression model trained successfully.")
```

## Generate Predictions

### Subtask:
Use the trained prediction model to make loan payback predictions on the preprocessed test dataset.


**Reasoning**:
I will use the trained Logistic Regression model to predict the probability of loan payback (positive class) on the preprocessed test dataset (`X_test_final`), as specified in the instructions, and store these probabilities.



```python
y_pred_proba = model.predict_proba(X_test_final)[:, 1]

print("Predicted probabilities for the positive class (loan paid back) on the test set:")
print(y_pred_proba[:5])
```

**Reasoning**:
With the predictions generated, the next step is to create the submission file by combining the test IDs and the predicted probabilities into a DataFrame and saving it in the specified format.



```python
submission_df = pd.DataFrame({'id': test_ids, 'loan_paid_back': y_pred_proba})

# Save the submission file
submission_df.to_csv('submission.csv', index=False)

print("Submission file created successfully.")
print(submission_df.head())
```
